### WHICH CAR BRAND HAS BEEN SOLD HIGHEST IN USED CAR?

```python
plt.figure(figsize=(16,16))
df['brand'].value_counts().plot(kind="bar")
plt.title("CAR BRAND")
plt.xlabel("BRAND")
plt.ylabel("NUMBER OF CARS")
plt.legend()
plt.show()
```

#### ---> THROUGH THIS BAR GRAPH WE COME TO KNOW THAT MARUTI BRAND CARS HAS BEEN SOLD HIGHEST AFTER THAT HYUNDAI

### NOW CHECKING WHICH CAR MODEL HAS BEEN SOLD HIGHEST?

```python
plt.figure(figsize=(22,32))
df['car_name'].value_counts().plot(kind="barh")
plt.title("CAR MODELS")
plt.xlabel("MODELS")
plt.ylabel("NUMBER OF CARS")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
